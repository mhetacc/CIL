```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
torch.manual_seed(0)
```

```python
x = torch.rand(10, requires_grad=True)
target = torch.ones(10)
```

```python
#optimizer = torch.optim.SGD([x], lr=0.01)
# optimizer = torch.optim.SGD([x], lr=0.01, momentum=0.9)
optimizer = torch.optim.Adam([x], lr=0.01)

loss_fn = nn.MSELoss()

losses = []

for epoch in range(150):
  optimizer.zero_grad()
  loss = loss_fn(target, x)
  loss.backward()
  optimizer.step()

  losses.append(loss.item())

  if epoch % 10 == 0:
    print(f'Epoch {epoch}, loss: {loss.item():.3f}  mean: {torch.mean(x).item():.3f}')
```

```python
print(x)
```

```python
plt.semilogy(losses)
plt.show()
```
