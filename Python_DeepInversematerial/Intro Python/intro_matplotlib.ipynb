```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-np.pi, np.pi, 250)
y1 = np.cos(x)
y2 = np.sin(x)

plt.plot(x,y1,'-', color='r')
plt.plot(x,y2,'--', color='k', linewidth=0.5)

plt.xlabel(r'$x$')
plt.ylabel(r'$y$')

plt.legend([r"$\cos(x)$",r"$\sin(x)$"])

plt.title('Trigonometric functions')
plt.grid()
plt.show()
```

```python
x = np.arange(1,9)
y = x**2
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6,6))
ax[0,0].plot(x, y, 'ko')
ax[0,1].plot(x, y, 'k-')
ax[1,0].plot(x, y, 'k-o')
ax[1,1].plot(x, y, 'k--o')

ax[1,1].set_title('Last example x')
plt.suptitle('Markers examples')

plt.show()
```

```python
N = 20000

x = np.random.normal(0, 1, (N, ))

plt.hist(x, bins=100)
plt.show()
```

```python
x = np.array(("a" , "a" , "a" , "b" , "a" , "b" ))
unique = np.unique(x)
count = [np.sum(x==el) for el in unique]

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(unique,count)
plt.title('Bar plot')

plt.subplot(1, 2, 2)
plt.pie(count,labels=unique)
plt.title('Pie chart')

plt.savefig('test.png')

plt.show()
```

```python
import skimage

img = skimage.data.astronaut()
print(img.shape)
```

```python
plt.figure(figsize=(5,5))
plt.imshow(img)
plt.show()
```

```python
R = img[:,:,0]
G = img[:,:,1]
B = img[:,:,2]

print(f'R channel shape is {R.shape}')
print(f'G channel shape is {G.shape}')
print(f'B channel shape is {B.shape}')

print(f'R channel max value is {np.max(R)}')
print(f'G channel max value is {np.max(G)}')
print(f'B channel max value is {np.max(B)}')
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(R, cmap=plt.cm.gray)
plt.colorbar(orientation='vertical')
plt.title('Red channel')

plt.subplot(1,3,2)
plt.imshow(G, cmap=plt.cm.gray)
plt.colorbar(orientation='vertical')
plt.title('Green channel')

plt.subplot(1,3,3)
plt.imshow(B, cmap=plt.cm.gray)
plt.colorbar(orientation='vertical')
plt.title('Blue channel')

plt.show()
```
