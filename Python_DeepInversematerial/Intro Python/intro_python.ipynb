```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello, my name is {self.name}.")
```

```python
p1 = Person("Alice", 25)
print(type(p1))
```

```python
p1.greet()
```

```python
class Student(Person):
    def __init__(self, name, age, grade):
        super().__init__(name, age)
        self.grade = grade

    def show_grade(self):
        print(f"{self.name} is in grade {self.grade}.")
```

```python
s1 = Student("Bob", 20, "A")
print(type(s1))

s1.greet()
s1.show_grade()
```
